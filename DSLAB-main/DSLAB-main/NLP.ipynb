```python
import numpy as n
import pandas as p
import matplotlib.pyplot as plt
import os
import pathlib
from PIL import Image
```

```python
import nltk

from nltk import word_tokenize
from nltk.probability import FreqDist
import urllib.request
import wordcloud
from wordcloud import WordCloud
%matplotlib inline
```

```python
nltk.download('punkt')
agatha=open('Downloads/Agatha.txt','r').read()
```

```python
print(agatha[100:200])
```

```python
words=word_tokenize(agatha)
print(len(words))
```

```python
fq=FreqDist(words)
fq.most_common(5)
```

```python
words_no_punc=[]
for w in words:
    if w.isalpha():
        words_no_punc.append(w)
print(len(words_no_punc))
```

```python
nltk.download("stopwords")
from nltk.corpus import stopwords
stop_words_list=stopwords.words("english")
print(len(stop_words_list))
```

```python
clean_words=[]
for w in words_no_punc:
    if w not in stop_words_list:
        clean_words.append(w)
print(len(clean_words))

fq=FreqDist(clean_words)
fq.most_common(5)
```

```python
clean_words_string=" ".join(clean_words)
wordcloud=WordCloud(background_color="orange").generate(clean_words_string)
plt.figure(figsize=(7,6))
plt.imshow(wordcloud)
plt.show()
```

```python
mask_image3=n.array(Image.open('Downloads/book.webp'))
```

```python
wordcloud=WordCloud(mask=mask_image3,background_color='white').generate(clean_words_string)
plt.figure(figsize=(12,4))
plt.imshow(wordcloud)
plt.show()
```
